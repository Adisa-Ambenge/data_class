```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import Image
%matplotlib inline
import os
```

```python
image_path = os.path.abspath("./paris/paris.jpg")
cb_img = cv2.imread(image_path)

if cb_img is None:
    print("Error: Unable to load the image.")
else:
    plt.imshow(cb_img, cmap='gray')
    plt.title("Grayscale Image")
    plt.show()
```

```python
cb_img_copy = cb_img.copy()
cb_img_copy[2000,2000] = 0
cb_img_copy[1900,1900] = 0
cb_img_copy[1950,1950] = 0
cb_img_copy[1980,1980] = 0

plt.imshow(cb_img_copy, cmap = 'gray')
plt.title("Grayscale Image")
plt.show()



```

```python
img_NZ_bgr = cv2.imread("./paris/paris.jpg",cv2.IMREAD_COLOR)
plt.imshow(img_NZ_bgr)
img_NZ_rgb = img_NZ_bgr[:,:,::-1]
plt.imshow(img_NZ_rgb)
```

```python
cropped_region = img_NZ_rgb[0:3000,3000:4500]
plt.imshow(cropped_region)
```

```python
img_NZ_rgb_flipped_horz = cv2.flip(img_NZ_rgb, 1 )
img_NZ_rgb_flipped_vert = cv2.flip(img_NZ_rgb,0)
img_NZ_rgb_flipped_both =cv2.flip(img_NZ_rgb, -1)


plt.figure(figsize =[18,5])
plt.subplot(151);plt.imshow(img_NZ_rgb_flipped_horz);plt.title ("Horizontal Flip")
plt.subplot(152);plt.imshow(img_NZ_rgb_flipped_vert);plt.title ("Vertical Flip")
plt.subplot(153);plt.imshow(img_NZ_rgb_flipped_both);plt.title ("Both Flip")
plt.subplot(154);plt.imshow(img_NZ_rgb);plt.title ("Original")
plt.subplot(155);plt.imshow(img_NZ_rgb_flipped_both);plt.title ("Both Flip")

```
